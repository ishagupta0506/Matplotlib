```python
gas = pd.read_csv('./gas_prices.csv')
gas
```

```python
plt.plot(gas.Year,gas.USA)
plt.plot(gas.Year,gas.Canada)
plt.plot(gas['Year'],gas['South Korea'])
plt.show()
```

```python
plt.plot(gas.Year,gas.USA, 'b.-')
plt.plot(gas.Year,gas.Canada)
plt.plot(gas['Year'],gas['South Korea'])
print(gas.Year[::3])
plt.xticks(gas.Year[::3])
plt.legend()
plt.show()
```

```python
for country in gas:
    if country != 'Year':
        plt.plot(gas.Year,gas[country],marker='_')
print(gas.Year[::3])
plt.xticks(gas.Year[::3])
plt.legend()
plt.show()
```

```python
bins=[90,200,275]
plt.hist(gas,bins=bins)
plt.show()
```
